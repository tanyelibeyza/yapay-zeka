```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import seaborn as sns

# Iris veri setini yükle
iris = load_iris()
X = iris.data  # Özellikler
y = iris.target  # Gerçek etiketler (gerçek kümeler)

# K-means algoritmasını uygula
kmeans = KMeans(n_clusters=3, random_state=42)  # 3 küme bekliyoruz (setosa, versicolor, virginica)
kmeans.fit(X)

# Kümeleri ve küme merkezlerini al
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

# Sonuçları görselleştir
# Özellikler: Sepal uzunluğu ve genişliği (ilk 2 özellik)
plt.figure(figsize=(8,6))
sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=labels, palette="Set2", s=100, alpha=0.6, edgecolor='black')

# Küme merkezlerini işaretle
plt.scatter(centroids[:, 0], centroids[:, 1], marker='X', s=200, c='red', label='Küme Merkezleri')

# Grafik ayarları
plt.title("K-means Kümeleme (Iris Veri Seti)")
plt.xlabel("Sepal Uzunluğu")
plt.ylabel("Sepal Genişliği")
plt.legend()
plt.show()

# Küme etiketlerini ve merkezlerini yazdır
print("Küme Etiketleri:", labels)
print("Küme Merkezleri:\n", centroids)
```
