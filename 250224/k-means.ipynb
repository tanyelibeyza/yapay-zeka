```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import seaborn as sns

# Iris veri setini ykle
iris = load_iris()
X = iris.data  # zellikler
y = iris.target  # Gerek etiketler (gerek kmeler)

# K-means algoritmasn uygula
kmeans = KMeans(n_clusters=3, random_state=42)  # 3 kme bekliyoruz (setosa, versicolor, virginica)
kmeans.fit(X)

# Kmeleri ve kme merkezlerini al
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

# Sonular grselletir
# zellikler: Sepal uzunluu ve genilii (ilk 2 zellik)
plt.figure(figsize=(8,6))
sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=labels, palette="Set2", s=100, alpha=0.6, edgecolor='black')

# Kme merkezlerini iaretle
plt.scatter(centroids[:, 0], centroids[:, 1], marker='X', s=200, c='red', label='Kme Merkezleri')

# Grafik ayarlar
plt.title("K-means Kmeleme (Iris Veri Seti)")
plt.xlabel("Sepal Uzunluu")
plt.ylabel("Sepal Genilii")
plt.legend()
plt.show()

# Kme etiketlerini ve merkezlerini yazdr
print("Kme Etiketleri:", labels)
print("Kme Merkezleri:\n", centroids)
```
